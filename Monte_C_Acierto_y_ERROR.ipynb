# Metodo de Monte Carlo de Acierto y Error para $\sqrt{tan^{-1} (x)} $

```python
import matplotlib.pyplot as plt
import numpy as np
import random as rd

```

Este metodo nos sirve para aproximar el area bajo una curva mediante los puntos aleatorios como se muestra en el siguiente apartado de nuestro codigo. Tomamos la funcion dada:

```python
# Definimos la función
def f(x):
    return np.sqrt(np.arctan(x))
```

Y comenzamos a calcular/obtener los puntos aleatorios en los ejes x,y asegurandonos que este debajo de la funcion


```python
def montecarlo(n):
    x_rand = np.random.uniform(0, 1, n)  # Valores aleatorios en [0,1]
    y_rand = np.random.uniform(0, f(1), n)  # Valores aleatorios en [0, f(1)]
    aciertos = np.sum(y_rand <= f(x_rand))  # Contamos los puntos bajo la curva
    estimacion = (aciertos / n) * (1 * f(1))
    return estimacion
```

```python
# Número de muestras
n = 300

resultado = montecarlo(n)
print(f"El resultado final de nuestra aproximación es: {resultado:.4f}")
```

```python
#Comenzamos una funcion donde generaremos una lista con los valores obtenidos en la funcion anterior con MC

def resultado(n):
    return [montecarlo(i) for i in range(1, n+1)] # esto nos da diferentes muestras que nos ayudara para graficarlo
```

```python
estimaciones = resultado(n)

plt.figure(figsize=(8, 5))
plt.plot(range(1, n+1), estimaciones, label='Estimación de la probabilidad', color='b')
plt.axhline(y=estimaciones[-1], color="red", linestyle="--", label=f'Última estimación: {estimaciones[-1]:.4f}')
plt.xlabel('Número de lanzamientos')
plt.ylabel('Estimación de la integral')
plt.legend()
plt.grid(True)
plt.show()
```
