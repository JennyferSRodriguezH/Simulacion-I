 Implemente la Prueba de Rachas para números aleatorios. Como prueba, use los datos del ejercicio realizado en la clase correspondiente.


```python
import random

def lanzar_monedas(num_lanza):
    """Simularemos el lanzamiento de una moneda y calcularemos la frecuencia de aguilas y soles."""

    secuencia = [] #almacenamos la secuencia de los lanzamientos
    cont_aguila = 0
    cont_sol = 0

    for _ in range(num_lanza): #repetiremos el proceso de acuerdo al num de lanzamientos
        resultado = random.random() #generamos un numero aleatorio
        if resultado < 0.5:
            secuencia.append("A")  # A para diferenciar Aguila
            cont_aguila += 1 #aumenta el contador de A
        else:
            secuencia.append("S")  # S para diferenciar a Sol
            cont_sol += 1

    prob_aguila = cont_aguila / num_lanza if num_lanza > 0 else 0 #con estas lineas calculamos las probabilidades
    prob_sol = cont_sol / num_lanza if num_lanza > 0 else 0   # para ambas caras de la moneda evitando
                                                              # dividir por cero
    return prob_aguila, prob_sol

def realizar_exp(N_experimentos): #repetiremos los lanzamientos
    """Realiza múltiples experimentos y almacena los resultados."""
    resultados = []

    for exp in range(1, N_experimentos + 1):  # Empiezamos desde 1 para evitar el caso de 0 lanzamientos
        prob_aguila, prob_sol = lanzar_monedas(exp) #llamamos la primer funcion
        resultados.append((exp, prob_aguila, prob_sol)) #con append guardamos los lanzamientos y las probabilidades

    return resultados

# Para este ejemplo usaremos la simulación con 10 experimentos
N_experimentos = 10
datos_exp = realizar_exp(N_experimentos) #llamamos la funcion con el valor de los 10 exp

# Mostraremos los resultados tomando en cuenta sus primeros 4 decimales
for exp, prob_aguila, prob_sol in datos_exp:
    print(f"Experimento {exp}: P(A) = {prob_aguila:.4f}, P(S) = {prob_sol:.4f}")
```
