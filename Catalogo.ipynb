# Teorema del Limite Central

El **teorema del Limite Central** es uno de los resultados fundamentales en la estadistica y la teoria de probabilidad. Se establece lo siguiente: \\
Sea $X_1 , X_2 \ldots, X_n$ una secuencia de variables aleatorias independientes e identicamente distribuidas, con una media $\mu$ y varianza finita $\sigma^2$. Entonces cuando $n \to \infty $, la distribucion de la variable aleatoria estandarizada:
$$
Z_n = \frac{\overline{X}_n - \mu}{\sigma / \sqrt{n}}
$$
converge en una distribucion normal estandar:
$$
Z_n \xrightarrow{d} \mathcal{N}(0, 1)
$$

# Distribuciones Continuas


Estas distribuciones se usan para variables que pueden tomar valores infinitos en un intervalo dado y la variable aleatoria puede tomar infinitos valores reales definidos en el intervalo. El valor exacto de las probabilidades suman cero.

## Distribucion Uniforme



*   Nomenclatura Estandar \
$$ X \sim \mathcal{U}(a, b)$$

*   Funcion de Densidad
$$
f(x)= \begin{cases}
        \frac{1}{b-a} &\ \text{para} \ a \leq x \leq b \\
        0 &\ \text{para} \ x < a \  \text{o} \ x > b
        \end{cases}
$$
*   Funcion de Distribucion Acumulada
$$
F(x) =\begin{cases}
    0, & x < a \\
    \frac{x - a}{b - a}, & a \leq x \leq b \\
    1, & x > b \end{cases}
$$
*   Media
$$
\mu = E[X] = \frac{a+b}{2}
$$
*   Varianza
$$
Var(X)= \frac{(b-a)^2}{12}
$$
*   Desviacion Estandar
$$
\sigma = \sqrt{Var(X)} = \frac{b-a}{\sqrt{12}}
$$



* Pseudocódigo



```
Entrada: lista de n elementos
          a y b de la distribución
Salida: valores aleatorios de X ~ U(a, b)

Iniciamos un ciclo hasta n
Generamos R con valores aleatorios
Calculamos a X[i] ← a + (b - a) * R

Return X
```

*   Código con histograma:




````python
import numpy as np
import matplotlib.pyplot as plt
import random as rd
````

````python
def uniforme(a,b,n):
  X = []
  for _ in range(n):
    R = np.random.rand()
    x = a+(b-a)* R
    X.append(x)
  return X
````

````python
resultado = uniforme(2,8,1000)
````

````python
plt.hist(resultado, bins = 20, color='pink', edgecolor='black')
plt.xlabel('X')
plt.ylabel('Densidad')
plt.title('Histograma Distribucion Uniforme')
plt.show()
````

## Distribucion Triangular      



*   Nomenclatura Estandar
$$
X \sim Tri(a, c, b)
$$
*   Funcion de Densidad.
$$
\begin{cases}
        \ \ \ \ \ \ 0 \ \ \ \ \ \ \ \ \ \text{para} \ \ x < a\\
        \frac{2(x-a)}{(b-a)(c-a)} \ \ \text{para} \ \ a \leq x <c\\
        \ \ \ \ \frac{2}{b-a} \ \  \ \ \ \ \ \ \text{para} \ \ x=c \\
        \frac{2(b-c)}{(b-a)(b-c)} \ \ \ \text{para} \ \ c<x \leq b \\
        \ \ \ \ \ \ 0 \ \ \ \ \ \ \ \ \ \ \ \text{para}\ \ b<x
    \end{cases}
$$
*   Funcion de Distribucion Acumulada
$$
\begin{cases}
     \ \ \ \ \ \ 0 \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \text{para} \ \ \ \ \ \ \ \ \   x \ \leq a \\
      \frac{(x-a)^2}{(b-a)(c-a)} \ \ \ \ \ \ \ \ \ \ \ \text{para}\ \ \ \  \ \ a<x \leq c \\
     1-\frac{(b-x)^2}{(b-a)(b-c)} \ \ \ \ \ \text{para} \ \ \ \ \ c<x<b \\
     \ \ \ \ \ \ \ \ 1 \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \text{para} \ \ \ \ \ \ \ \ \ \ b\leq x
    \end{cases}
$$
*   Media
$$
 \mu = E[X] = \frac{a+b+c}{3}
$$
*   Varianza
$$
Var(X) = \frac{a^2 + b^2+c^2 -ab-ac-bc}{18}
$$
*   Desviacion Estandar
$$
\sigma = \sqrt{Var(X)} = \sqrt{\frac{a^2 + b^2+c^2 -ab-ac-bc}{18}}
$$

*   Pseudocódigo



```
Entrada: lista de n elementos
         m parámetro de la distribución Triang(0, 1, m)
Salida: valores aleatorios de X

Para i desde 1 hasta n hacer:
    Generar R ← número aleatorio en [0, 1]
    Si R ≤ m entonces:
        X[i] ← sqrt(m * R)
    Si no:
        X[i] ← 1 - sqrt((1 - m) * (1 - R))

Retornar X
```






*   Codigo con histograma



````python
import numpy as np
import matplotlib.pyplot as plt
import random as rd
````

````python


def triangular(n):
    X = []
    m = 0.6
    for _ in range(n):
        R = np.random.rand()
        if R < m:
            x = np.sqrt(m * R)
        else:
            x = 1 - np.sqrt((1 - m) * (1 - R))
        X.append(x)
    return X
````

````python
resultado = triangular(1000)
````

````python
plt.hist(resultado, color='pink', edgecolor='black')
plt.xlabel('X')
plt.ylabel('Densidad')
plt.title('Histograma Distribucion Triangular ')
plt.show()
````

## Distribucion Exponencial

$$
X \sim Exp(\lambda)
$$

$$
f(x) = \lambda e^{-\lambda x} , \ \ \ \ x \geq 0
$$
*   Funcion de Distribucion Acumulada

F(x) = 1-e^{-\lambda x}, \ \ \ \ x \geq 0

Var[X]=\frac{1}{\lambda^2}

## Distribucion m-Erlang



*   Nomenclatura Estandar
$$
 X \sim Erlang(n, \lambda)
$$
*   Funcion de Densidad.
$$
f(x)= \frac{\lambda (\lambda x)^{n-1}e^{-\lambda x}}{(n-1)!}, \ \ \ x\geq 0
$$
*   Funcion de Distribucion Acumulada
$$
F(x)= 1- \sum_{k=0}^{n-1} \frac{(\lambda x)^{k} e^{-\lambda x}}{k!}
$$
*   Media
$$
\mu = E[X] = \frac{m}{\lambda}
$$
*   Varianza
$$Var(X) = \frac{m}{\lambda^2}
$$
*   Desviacion Estandar
$$
\sigma = \frac{\sqrt{m}}{\lambda}
$$

```
Entrada:

Salida:
    X

Producto ← 1

Para i desde 1 hasta m hacer:
    Generar R ← número aleatorio en [0, 1]
    Producto ← Producto * R

X ← -(b - m) * ln(Producto)

Retornar X


```

````python
import numpy as np
import matplotlib.pyplot as plt
import random as rd
````

````python
def earlang(n):
  X =[]
  m = 3
  b = 5
  for _ in range(n):
    producto = 1
    for i in range(m):
      R = np.random.rand()
      producto *= R
    x = -(b-m)*np.log(producto)
    X.append(x)
  return X
````

##  Distribucion de Gamma



*   Nomenclatura Estandar
$$
f(x)= \Gamma (\alpha, \lambda)
$$
*   Funcion de Densidad.
$$
f(x)= \frac{\lambda}{\Gamma(\alpha)}(\lambda x)^{\alpha -1} e^{-\lambda x} \ \ \ para \ x>0
$$
donde:
$$
\Gamma (\alpha)= \int_{0}^{\infty} t^{\alpha -1} e^{-t} dt
$$

*   Funcion de Distribucion Acumulada
$$
F(x)= \frac{1}{\Gamma(\alpha)}\int_{0}^{x /\beta}t^{\alpha -1}e^{-t}dt = P(\alpha, x/\beta)
$$
*   Media
$$
\mu = E[X] = \alpha \beta
$$
*   Varianza
$$
Var(X)= \alpha \beta^2
$$
*   Desviacion Estandar
$$
\sigma = \sqrt{\alpha} \cdot\beta
$$

densidad Gamma.png


* Grafica Acumulativa:

acumulada Gamma.png

* Pseudocódigo


```
Entrada:
    alpha y beta (parámetros de la distribución)

Salida:
    valores aleatorios X

Si alpha > 1 entonces:
    d ← alpha - 1/3
    c ← 1 / sqrt(9 * d)

    Repetir:
        Generar R ← aleatorio uniforme en [0, 1]
        Generar Z ← aleatorio normal estándar N(0, 1)
        V ← (1 + c * Z)^3

        Si V > 0 entonces:
            X ← d * V
            A ← ln(R)
            B ← 0.5 * Z^2 + d - d * V + d * ln(V)
            
            Si A < B entonces:
                Retornar X * beta

Sino:  (0 < alpha ≤ 1)
    Generar X ← usando el caso alpha > 1 con alpha + 1
    Generar R ← aleatorio uniforme en [0, 1]
    Retornar X * R^(1 / alpha) * beta

```



* Codigo con Histograma

````python
import numpy as np
import matplotlib.pyplot as plt
import random as rd
````

````python
import numpy as np
import matplotlib.pyplot as plt

def gamma(n, alpha, beta):
    lista = []
    for _ in range(n):
        if alpha >= 1:
            d = alpha - 1/3
            c = 1 / np.sqrt(9 * d)
            while True:
                R = np.random.randn()
                V = (1 + c * R) ** 3
                if V > 0:
                    U = np.random.rand()
                    if np.log(U) < 0.5 * R**2 + d * (1 - V + np.log(V)):
                        lista.append(d * V * beta)
                        break
        else:

            while True:
                d = (alpha + 1) - 1/3
                c = 1 / np.sqrt(9 * d)
                R = np.random.randn()
                V = (1 + c * R) ** 3
                if V > 0:
                    U1 = np.random.rand()
                    if np.log(U1) < 0.5 * Z**2 + d * (1 - V + np.log(V)):
                        G = d * V
                        U2 = np.random.rand()
                        x = G * (U2 ** (1 / alpha))
                        lista.append(x * beta)
                        break
    return lista
````

## Distribucion Weibull



*   Nomenclatura Estandar
$$
X \sim Weibull(\alpha, \lambda)
$$
*   Funcion de Densidad.
$$
f(x) = \begin{cases}
        \frac{k}{\lambda}(\frac{x}{\lambda})^{k-1} e^{-(x/\lambda)^k}, \ \ \  \ \ \ x \geq 0 \\
        \ \ \ \ \ \ \ \ \ \ \ \ 0 \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ x<0 \\
        \end{cases}
$$
o tambien:
$$f(x)= \lambda\alpha(\lambda x)^{\lambda - 1}e^{-(\lambda x)^{\alpha}}, \ \ x> 0
$$
*   Funcion de Distribucion Acumulada
$$
F(x)=1-e^{-(\lambda x)^{\alpha}}, \ \ \ x\geq 0
$$
*   Media
$$
\mu = E[X] = \frac{1}{\lambda} \Gamma (1+ \frac{1}{\alpha})
$$
*   Varianza
$$
Var(X)= \frac{1}{\lambda^2}[\Gamma (1+\frac{2}{\alpha}) - \Gamma^2(1 + \frac{1}{\alpha})]
$$
*   Desviacion Estandar
$$
\sigma = \sqrt{Var(X)}
$$

* Pseudocódigo


```
Entrada:
    n: cantidad de valores
    alpha: parámetro de forma (α)
    lambda: parámetro de escala (λ)

Salida:
    Lista de valores aleatorios X

Generar n números aleatorios R con distribución U(0,1)

Para cada R:
    X ← λ * (-ln(1 - R))^(1 / α)

Retornar X

```



* Código con Histograma

````python
import numpy as np
import matplotlib.pyplot as plt
import random as rd
````

````python
def weibul(n):
  X =[]
  alpha = 1.5
  lam = 2
  for _ in range(n):
    R = np.random.rand()
    x = lam*(-np.log(1-R))**(1/alpha)
    X.append(x)

  return X
````

## Distribucion Normal



*   Nomenclatura Estandar
$$
X \sim \mathcal{N}(\mu, \sigma^2)
$$
*   Funcion de Densidad.
$$
f(x)= \frac{1}{\sigma \sqrt{2 \pi}} e^{-\frac{(x-\mu)^2}{2\sigma ^2}}
$$
*   Funcion de Distribucion Acumulada
$$
F(x)= \frac{1}{2}[1+ erf (\frac{x - \mu}{\sigma  \sqrt{2}})]
$$
donde  **erf** es la función de error.
*   Media
$$
\mu = E[X] = \frac{1}{\lambda} Γ(1 + \frac{1}{\alpha})
$$
*   Varianza
$$
\sigma^2 = Var(X) = \frac{1}{\lambda^2} [\Gamma (1 + \frac{2}{α}) - \Gamma^2 (1+ \frac{1}{\alpha})]
$$
*   Desviacion Estandar
$$
\sigma = \sqrt{Var(X)}
$$

*Pseudocódigo:
Para esta distribucion podemos implementar el *Método de Box-Muller*

```
Entrada:
    n: número de valores aleatorios a generar

Salida:
    Listas de valores aleatorios X y Y

Para i desde 1 hasta n/2 hacer:
    Generar R1 y R2 de la forma U(0,1)
    Calcular R ← sqrt(-2 * ln(R1))
    Calcular θ ← 2 * π * R2
    Calcular X ← R * cos(θ)
    Calcular Y ← R * sin(θ)
    
Guardar X e Y en sus respectivas listas

Retornar listas X, Y

```



* Codigo con Histograma

````python
import numpy as np
import matplotlib.pyplot as plt
import random as rd
````

````python
def boxmuller(n):
    X = []
    Y = []
    for i in range(int(n/2)):
        R1 = np.random.rand()
        R2 = np.random.rand()
        R= np.sqrt(-2*np.log(R1))
        theta = 2*np.pi*R2
        x = R * np.cos(theta)
        y = R * np.sin(theta)

        X.append(x)
        Y.append(y)

    return X,Y
````

````python
resultado = boxmuller(1000)
````

````python
X, Y = boxmuller(1000)
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.hist(X, bins=30, color='pink', edgecolor='black', density=True)
plt.title('Histograma de $X \\sim \\mathcal{N}(0,1)$')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.hist(Y, bins=30, color='pink', edgecolor='black', density=True)
plt.title('Histograma de $Y \\sim \\mathcal{N}(0,1)$')
plt.grid(True)

plt.tight_layout()
plt.show()
````

## Distribucion Beta



*   Normenclatura Estandar
$$
X \sim Beta(\alpha, \beta)
$$
*   Funcion de Densidad. \\
Para $ x \in (0,1)$
$$
f(x)=\frac{x^{\alpha -1}(1-x)^{\beta-1}}{B(\alpha, \beta)}
$$
donde $B(\alpha, \beta)$ es la función beta: \\
$$
B(\alpha, \beta) = \int_{0}^1 t^{\alpha-1}(1-t)^{\beta-1} dt = \frac{\Gamma(\alpha) \Gamma(\beta)}{\Gamma(\alpha + \beta)}
$$

*   Funcion de Distribucion Acumulada
$$
F(x)= I_{x}(\alpha, \beta)
$$
*   Media
$$
\mu = E[X] = \frac{\alpha}{\alpha + \beta}
$$
*   Varianza
$$
 Var(X)= \frac{\alpha \beta}{(\alpha + \beta)^2(\alpha+\beta+1)}
$$
*   Desviacion Estandar
$$
σ = \sqrt{\frac{\alpha \beta}{(\alpha + \beta)^2(\alpha+\beta+1)}}
$$

densidad beta.png

* Grafica Acumulada:
acumulada beta.png

* Pseudocódigo:


```
Entrada:
    n: número de valores a generar
    α > 0
    β > 0

Salida:
    Lista de valores aleatorios Y

Para i desde 1 hasta n hacer:
    Generar R1 ∼ Gamma(α, 1)
    Generar R2 ∼ Gamma(β, 1)
    Calcular Y ← R1 / (R1 + R2)
    
Guardar Y en la lista

Retornar lista Y

```



* Codigo con Histograma

````python
import numpy as np
import matplotlib.pyplot as plt
import random as rd
````

````python

def beta(n, alpha=2, beta=5):
    Y = []
    for _ in range(n):
        R1 = np.random.gamma(shape=alpha, scale=1.0)
        R2 = np.random.gamma(shape=beta, scale=1.0)
        y = R1 / (R1 + R2)
        Y.append(y)
    return Y
````

````python
resultado=beta(1000)
````

````python
plt.hist(resultado, bins = 20, color='pink', edgecolor='black')
plt.xlabel('Y')
plt.ylabel('Densidad')
plt.title('Histograma Distribucion de Beta')
plt.show()
````

## Distribución Pearson Tipo V   



*   Nomenclatura Estandar: \\
Sea $$X \sim Pearson V (\alpha , \lambda)$$
donde: \\
$\alpha > 0 $: parámetro de forma. \\
$\lambda > 0 $: parámetro de escala. \\
*   Funcion de Densidad.
$$
f(x) = \frac{\lambda^\alpha}{\Gamma(\alpha)} x^{-\alpha - 1} e^{-\lambda / x}, \qquad x > 0
$$
*   Funcion de Distribucion Acumulada
$$F(x) = \frac{1}{\Gamma(\alpha)} \int_{\lambda / x}^{\infty} t^{\alpha - 1} e^{-t} \, dt = \frac{\Gamma\left(\alpha, \frac{\lambda}{x}\right)}{\Gamma(\alpha)}
$$
donde $\Gamma(s, x)$es la función gamma incompleta superior.

*   Media
$$ \mu = E[X] = \dfrac{\lambda}{\alpha - 1}
    \quad \ si  \ \  \alpha > 1
$$
*   Varianza
$$
 Var(X) = \dfrac{\lambda^2}{(\alpha - 1)^2(\alpha - 2)} \quad si \ \ \alpha > 2
 $$
*   Desviacion Estandar
$$
\sigma = \frac{\lambda}{(\alpha - 1)\sqrt{\alpha - 2}} \quad \text{si } \alpha > 2
$$

* Pseudocódigo:



```
Entrada:
    n: número de muestras
    a₁, a₂: parámetros de forma de las distribuciones Gamma
    b: parámetro de escala de Y₁

Salida:
    Lista X de n muestras

Inicializar lista vacía X

Para i desde 1 hasta n hacer:
    Generar Y₁ ∼ Gamma(a₁, b)
    Generar Y₂ ∼ Gamma(a₂, 1), independiente de Y₁
    Calcular Xᵢ ← Y₁ / Y₂
    Guardar Xᵢ en la lista X

Retornar lista X

```



* Código con Histograma

````python
import numpy as np
import matplotlib.pyplot as plt
import random as rd
````

````python


def pearsonV(n, alpha=5, beta=2):
    X = []
    for _ in range(n):
        # Primero generamos una muestra de Gamma(alpha, beta)
        Y = np.random.gamma(shape=alpha, scale=1/beta)  # nota: scale = 1/beta
        # Pearson tipo V: tomamos el inverso
        x = 1 / Y
        X.append(x)
    return X
````

## Distribución Pearson Tipo VI

Esta distribucion es un conjunto de distribuciones continuas diseñadas por Karl Pearson paa modelar diferentes tipos de distribuciones. En este caso nos enfocaremos en la distribucion tipo **VI**



*   Nomenclatura Estandar
Sea $X \sim \text{Pearson VI}(\alpha, \beta, \lambda)$, donde: \\


  1.   $\alpha > 0$: parámetro de forma
  2.   $\beta > 0$: parámetro de forma
  3.   $\lambda > 0$: parámetro de escala.


*   Funcion de Densidad: \\
No existe una única formula para toda la familia, pero para este tipo de distribución, se puede presentar como: \\
$$
f(x) = \frac{\lambda^{\alpha} x^{\alpha - 1}}{B(\alpha, \beta)(1 + \lambda x)^{\alpha + \beta}}, \quad x > 0
$$
donde $B(\alpha, \beta)$ es la función beta de Euler: \\
$$
B(\alpha, \beta) = \int_0^1 t^{\alpha - 1} (1 - t)^{\beta - 1} dt
$$

*   Funcion de Distribucion Acumulada: \\
Esta distribucion no tiene forma cerrada en genral. Ya que se requiere una integracion numerica o el uso de funciones especiales.Pero podemos usar la función beta incompleta regularizada:
$$
F(x) = I_{\frac{\lambda x}{1 + \lambda x}}(\alpha, \beta)
$$

*   Media: \\
Media:  Existe cuando $\beta > 1$: \\
$$
\mu = E[X] = \frac{1}{\lambda} \cdot \frac{\alpha}{\beta - 1}
$$

*   Varianza \\
Si $\beta > 2$: \\
$$
Var(X) = \frac{1}{\lambda^2} \cdot \frac{\alpha(\alpha + \beta - 1)}{(\beta - 2)(\beta - 1)^2}
$$

*   Desviacion Estandar: \\
Si $\beta > 2$: \\
$$
\sigma^2 = \frac{1}{\lambda^2} \cdot \sqrt{\frac{\alpha(\alpha + \beta - 1)}{(\beta - 2)(\beta - 1)^2}}
$$

* Pseudocodigo: \\
Tomando en cuenta a la distribución tipo VI usando el método de la transformada inversa, nuestro pseudocódigo seria:



```
Entrada:
    n: número de muestras
    a₁, a₂: parámetros de forma de la distribución Beta
    b: parámetro de escala

Salida:
    Lista X de n muestras

Inicializar lista vacía X

Para i desde 1 hasta n hacer:
    Generar U ∼ U(0, 1)
    Calcular v ← F_B⁻¹(U)  # Inversa de la Beta(a₁, a₂)
    Calcular Xᵢ ← (b * v) / (1 - v)
    Guardar Xᵢ en la lista X

Retornar lista X

```



* Código con histograma

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta
import random as rd
````

````python
def beta(a, b):
    # Generamos una variable Beta(a, b)
    R1 = np.random.gamma(shape=a, scale=1.0) # dentro de un intervalo
    R2 = np.random.gamma(shape=b, scale=1.0)
    return R1 / (R1 + R2)
````

````python
def pearsonVI(n, a1=2, a2=3, b=1.5):
    X = []
    for _ in range(n):
        v = beta(a1, a2)  #usamos la función anterior
        x = (b * v) / (1 - v)
        X.append(x)
    return X
````

##  Distribucion Jhonson



*   Normenclatura Estandar
La variable aleatoria $X$ sigue una distribución Johnson $S_U(\gamma, \delta, \xi, \lambda)$, definida mediante la transformación:
$$
Z = \gamma + \delta \sinh^{-1}\left(\frac{X - \xi}{\lambda}\right),
$$
donde $Z \sim N(0,1)$ y:
$\gamma \in \mathbb{R}$: parámetro de localización. \\
  $\delta > 0$: parámetro de forma \\
  $\xi \in \mathbb{R}$: parámetro de localización de $X$ \\
  $\lambda > 0$: parámetro de escala de $X$.

*   Funcion de Densidad.
$$
f_X(x) = \frac{\delta}{\lambda \sqrt{2\pi} \sqrt{1 + \left(\frac{x - \xi}{\lambda}\right )^2}} \exp\left(-\frac{1}{2} \left[\gamma + \delta \sinh^{-1}\left(\frac{x - \xi}{\lambda}\right) \right]^2 \right), \quad x \in \mathbb{R}.
$$
*   Funcion de Distribucion Acumulada
$$
F_X(x) = \Phi\left(\gamma + \delta \sinh^{-1}\left(\frac{x - \xi}{\lambda}\right)\right),
$$
donde $\Phi$ es la función de distribución acumulada de la normal estándar.
*   **No existen** fórmulas para la media y varianza.

```
Entrada: número de muestras n, parámetros a1, a2, g, b
Salida: arreglo X de n muestras de la distribución.

Inicializar una lista de X tamaño n

Iniciamos un bucle de i hasta n:
    Generar Zi ~ N(0,1)
    Calcular Yi = exp((Zi - a1)/a2)
    Calcular Xi = g + (b/2) * (Yi - 1/Yi)

Retornar X

```

````python
import numpy as np
import matplotlib.pyplot as plt
import random as rd
````

````python
def jhonson(n,a1,a2,g,b):
  X=[]
  for i in range(n):
    R = np.random.rand()
    Y=np.exp((R-a1)/a2)
    x = g+(b/2)*(Y-1/Y)
    X.append(x)
  return X
````

## Distribucion Cauchy Estandar



*   Nomenclatura Estandar
$$
X \sim Cauchy(x_0, \gamma)
$$
Donde: \\
 $x_0 \in R$ indica la localizacion. \\
 $\gamma > 0$: indica la escala.
*   Funcion de Densidad.
$$
f(x) = \frac{1}{\pi \gamma \left[ 1 + \left( \frac{x - x_0}{\gamma} \right)^2 \right]}, \quad x \in (-\infty, \infty)
$$
Para Cauchy Estandar con $x_0= 0$ y $\gamma=1$:
$$
f(x) = \frac{1}{\pi (1 + x^2)}
$$

*   Funcion de Distribucion Acumulada
$$
F(x) = \frac{1}{\pi} \arctan\left( \frac{x - x_0}{\gamma} \right) + \frac{1}{2}
$$
Para Cauchy Estandar:
$$
F(x) = \frac{1}{\pi} \arctan(x) + \frac{1}{2}
$$
*   Para esta distribución la media, varianza y desviación estándar no existen, ya que esta distribución no cuenta con momentos definidos.

Estas distribuciones se usan cuando la variable aleatoria toma valores contables y la suma de sus pribabilidades de manera individual suman 1.

## Distribución Uniforme  



*   Normenclatura Estandar
$$
X \sim \mathcal{U}(a,b)
$$
donde: \\
$
    \mathcal{X} \text{ es una variable aleatoria discreta.}\\
    \mathcal{a} \ \  \text{y} \ \ \mathcal{b}. \
    \ \ \text{son enteros con}\  a \  \leq b .  \\
\ \     \text{Los valores posibles que toma } \mathcal{X} a, \  a+1 \ , \ a+2,..,b .\\
    \ \text{Hay $n = b-a+1$ posibles valores igualmente probables. }
    $
*   Funcion de Densidad.
$$
P(\mathcal{X}=x) \frac{1}{b-a+1}, \ \ \ x=a,a+1,...,b
$$
*   Funcion de Distribucion Acumulada
$$
F(x)= \begin{cases}
    0 & \text{si } x < a \\\\
    \frac{|x| - a + 1}{b - a + 1} & \text{si } a \leq x \leq b \\\\
    1 & \text{si } x > b
\end{cases}
$$

*   Media
$$
\mu = \frac{a+b}{2}
$$
*   Varianza
$$
 \sigma^2= \frac{(b-a+1)^2-1}{12}
$$
*   Desviacion Estandar
$$
 \sigma = \sqrt{\frac{(b-a+1)^2-1}{12}}
$$

```
Entrada: lista de n elementos, constante j > 0
Salida: valores aleatorios de X

Iniciar un bucle de i hasta n:
    Generar U = (U_1, U_2, ..., U_j) con U_i ∼ U(0,1)
    Calcular X = suma de (j - i + 1) * U_i desde i = 1 hasta j

Retornar X

```



* Código e Histograma

````python
def disc_unif(n,j):
  X = []
  for _ in range(n):
        R = np.random.uniform(0, 1, j)
        x = sum((j - i) * R[i] for i in range(j))
        X.append(x)
  return X
````

## Distribucion Binomial Negativa

$$ X \sim NegBin(r,p) $$

$X \in \{r, r+1, r+2,... \}$ y \\
         $r \in N$ numero de éxitos deseados. \\

        $X$ numero de endayor necesarios para obtener $r$ exitos.

## Distribucion Poisson



*   Nomenclatura Estandar
$$  X \sim Poisson(\lambda)$$
Donde: \\
$$ X \in \{0,1,2,3,...\}$$
   $$\lambda > 0$$ es el numero esperado de eventos en un intervalo fijo.

*   Funcion de Densidad.
$$P(X = x) = \frac{e^{-\lambda} \lambda^x}{x!}, \quad x = 0, 1, 2, \dots$$
*   Funcion de Distribucion Acumulada \\
Esta distribución no tiene una forma cerrada, por lo cual se calcula de la manera siguiente:
$$F(x)= \sum_{k=0}^{x} \frac{e^{-\lambda}\lambda^k}{k!}$$
*   Media
$$\mu = \lambda $$
*   Varianza
$$\sigma^2 = \lambda $$
*   Desviacion Estandar
$$ \sigma = \sqrt{\lambda}$$

* Pseudocódigo



```
Entrada: lista de n elementos
         parámetro λ
Salida: valores aleatorios de X

Iniciar un ciclo de k=1 hasta n:
    a ← exp(-λ)
    b ← 1
    i ← 0
    While b ≥ a:
        Generar U ∼ U(0,1)
        b ← b * U
        If b ≥ a:
            i ← i + 1
    X_k ← i

Return X

```



* Código e Histograma

````python
import numpy as np
import matplotlib.pyplot as plt
import random as rd
````

````python
def poisson(n, lam):
    X = []
    for _ in range(n):
        a = np.exp(-lam)
        b = 1
        i = 0
        while b >= a:
            R = np.random.uniform(0, 1)
            b *= R
            if b >= a:
                i += 1
        X.append(i)
    return X
````

## Distribucion Gumbel



*   Normenclatura Estandar
$$ X \sim Gumbel(\mu, \beta)$$
 $\mu \in \mathcal{R}$: parametro de localizacion.
 $\beta > 0$: parametro de escala.
 $X \in (-\infty, \infty)$

*   Funcion de Densidad.
$$f(x) = \frac{1}{\beta} \exp\left( -\frac{x - \mu}{\beta} - e^{-\frac{x - \mu}{\beta}} \right), \quad x \in (-\infty, \infty)$$
*   Funcion de Distribucion Acumulada
$$ F(x) = \exp\left( -e^{-\frac{x - \mu}{\beta}} \right) $$
*   Media
$$Media = \mu + \beta \gamma $$

```
Entrada: n (número de muestras)
Salida: lista de muestras de rango n

Para i desde 1 hasta n hacer:
    1. Generar u ← número aleatorio uniforme en (0,1)
    2. Calcular x ← ln(-ln(1 - u))
    3. Guardar x en la lista de muestras

Fin



```




* Código e Histograma

````python
import random
import math
import matplotlib.pyplot as plt
````

````python
def gumbel(n):
  X = []
  for _ in range(n):
    R=random.uniform(0,1)
    x=math.log(-math.log(1-R))
    X.append(x)
  return X
````

````python
resultado = gumbel(1000)
````

````python
plt.hist(resultado, color='pink', edgecolor='black')
plt.title('Distribucion de Gumbel')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.show()
````

## Distribucion Pareto



*   Nomenclatura Estandar
$$  X \sim Pareto(x_m,\alpha) $$
Donde: \\
$x_m > 0$: es el valor minimo. \\
$\alpha>0$: indicador de forma.
*   Funcion de Densidad.
$$ f(x) = \begin{cases}
\frac{\alpha x_m^\alpha}{x^{\alpha + 1}}, & x \geq x_m \\
0, & x < x_m
\end{cases}
$$
*   Funcion de Distribucion Acumulada
$$F(x) = \begin{cases}
1 - \left( \frac{x_m}{x} \right)^\alpha, & x \geq x_m \\
0, & x < x_m
\end{cases}$$
*   Media
$$ \mu = \frac{\alpha x_m}{\alpha - 1}, \quad \alpha > 1 $$
*   Varianza
$$ \sigma = \frac{x_m^2 \alpha}{(\alpha - 1)^2 (\alpha - 2)}, \quad \alpha > 2 $$
*   Desviacion Estandar
$$ \sigma^2= \sqrt{\text{Varianza}} $$
