1.   Datos del problema:
    * $X$: Tiempo de llegada de la Persona 1, en minutos.
    * $Y$: Tiempo de llegada de la Persona 2, en minutos.
    * $D=|X-Y|$: Diferencia absoluta entre los tiempos de llegada.
    * Esto se  distribuyen de forma **uniforme continua** en el intervalo $[0,30]$, es decir:
                 $X∼U(0,30),$  $Y∼U(0,30)$

    * Se generarán pares de valores $(X_i, Y_i)$, y se calculará $D_i=|X_i-Y_i|$
    * Tendremos un numero de n=100,000 repeticiones.



```python
import numpy as np
import matplotlib.pyplot as plt


n = 100000 # Número de simulaciones

# Generamos las llegadas de las personas
x = np.random.uniform(0, 30, n)  # Persona 1
y = np.random.uniform(0, 30, n)  # Persona 2

# Calculamos las diferencias absolutas con la formula ya especificada
D = np.abs(x - y)

# Contar cuántas diferencias cumplen la condición
eventos_favorables = D <= 10

# Probabilidad estimada
probabilidad = np.mean(eventos_favorables)
print(f"Probabilidad por simulación: {probabilidad:.4f}")
```

Graficamos lo anterior donde:


1.   Puntos azules: espera $\leq$ 10
2.   Puntos rojos: espera $>$ 10



```python
plt.scatter(x, y, c=eventos_favorables, cmap='bwr', s=1, alpha=0.3)
plt.xlabel('Persona 1')
plt.ylabel('Persona 2')
plt.title('Llegadas y espera con un maximo de 10 minutos')
plt.grid(True)
plt.show()
```
