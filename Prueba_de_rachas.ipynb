```python
import numpy as np
import scipy.stats as stats
from tabulate import tabulate  #edicionde las tablas

# Datos de la muestra
muestra = np.array([
    0.0449, 0.1733, 0.5746, 0.049, 0.8406, 0.8349, 0.92, 0.2564,
    0.6015, 0.6694, 0.3972, 0.7025, 0.1055, 0.1247, 0.1977, 0.0125,
    0.63, 0.2531, 0.8297, 0.6483, 0.6972, 0.9582, 0.9085, 0.8524,
    0.5514, 0.0316, 0.3587, 0.7041, 0.5915, 0.2523, 0.2545, 0.3044,
    0.0207, 0.1067, 0.3857, 0.1746, 0.3362, 0.1589, 0.3727, 0.4145
])

# Determinamos 0 si x_i < x_i-1, o 1 si x_i >= x_i-1
comparaciones = np.array([1 if muestra[i] >= muestra[i - 1] else 0 for i in range(1, len(muestra))])

#Contamos las rachas
def contar_rachas(S):
    rachas = 1  # Empezar con la primera racha
    for i in range(1, len(S)):
        if S[i] != S[i - 1]:  # Si hay cambio de valor (de 1 a 0 o de 0 a 1)
            rachas += 1
    return rachas

# Llamamos a la función para contar las corridas
Co = contar_rachas(comparaciones)

# Calculamos el valor esperado del número de corridas (mu)
mu = ((2 * len(muestra)) - 1) / 3

# Calculamos la varianza del número de corridas (sigma2)
sigma2 = ((16 * len(muestra)) - 29) / 90

# Calculamos el valor estadístico z0
z0 = (Co - mu) / np.sqrt(sigma2)


print(f"Número de corridas observadas (Co): {round(Co, 3)}")
print(f"Valor esperado de corridas (mu): {round(mu, 3)}")
print(f"Varianza del número de corridas (sigma2): {round(sigma2, 3)}")
print(f"Valor estadístico (z0): {round(z0, 3)}")


# Convertimos las listas en una tabla
tabla_resultados = np.vstack((muestra[1:], comparaciones)).T

headers = ["xi", "Corridas"]
print("\nTabla de corridas observadas")
print(tabulate(tabla_resultados, headers=headers, tablefmt="grid"))


# El intervalo de aceptación está dado por  -z_{alpha/2} <= z0 <= z_{alpha/2}
def valor_Z(alpha):
    # Calculamos Z_alpha/2 utilizando la distribución normal estándar
    return stats.norm.ppf(1 - alpha / 2)

# Definir el nivel de confianza (ejemplo: 95%)
nivel_confianza = 0.95

# Convertir el nivel de confianza en el valor alpha
alpha = 1 - nivel_confianza

# Calculamos el valor Z correspondiente
Z_alpha_2 = valor_Z(alpha)



# Mostrar el nivel de confianza y el valor de Z
print(f"\nNivel de confianza seleccionado: {nivel_confianza*100}%")
print(f"Para un nivel de aceptación del {nivel_confianza*100}%, Z_alpha/2 es: {round(Z_alpha_2, 3)}")

# Definimos el intervalo de aceptación
if -Z_alpha_2 <= z0 <= Z_alpha_2:
    print("\nNo se rechaza la hipótesis nula. ")
else:
    print("\nSe rechaza la hipótesis nula. ")
```
