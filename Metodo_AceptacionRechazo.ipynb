```python
import matplotlib.pyplot as plt
import random as rd
import numpy as np
```

```python
a=0
b=1
c=3
M=1
```

Para una distribucion triangular usamos las siguientes funciones:


```python
def f(x):
  if a<=x<=b:
    return 2*(x-a)/((c-a)*(b-a))
  elif b < x <=c:
    return -2*(x-c)/((c-a)*(c-b))
  else:
    return 0
```

```python
def aceptacion(n):
  Lista=[]
  for _ in range(n):
     R1= rd.random()
     R2= rd.random()
    #x = a+ (b-a)*R1
     x = a+ (c-a)*R1 #se modifica la formula para la dist. triang. ya que se quiere cubrir todo el dominio
     if R2 <= f(x)/M:
      Lista.append(x)

  return Lista
```

```python
C = aceptacion(5000)
```
