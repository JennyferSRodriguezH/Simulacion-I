Importamos las librerias necesarias para nuestro programa

```python
import matplotlib.pyplot as plt
import numpy as np
import math
import random as rd
```

Definimos los valores de **a,b,c**


```python
a=1
b=3
c=5
```

Definimos nuestra funcion inversa

```python
def inversa(n):
  x=[]
  for i in range(n):
    R= rd.random()
    if R<0.5:
      x.append(a+math.sqrt(R*(b-a)*(c-a)))
    else:
      x.append(c-math.sqrt((1-R)*(c-a)*(c-b)))
  return x
```

```python
inv = inversa(10000)
print(inv)
```

Graficamos el histograma de la funcion inversa

```python
plt.figure(figsize=(10,6))
plt.hist(inv, bins=100, density=True, alpha=0.6)
plt.show()
```

Calculamos la media y la varianza en los siguientes codigos


```python
med = np.mean(inv)
print(med)
```

```python
var = np.var(inv)
print(var)
```
