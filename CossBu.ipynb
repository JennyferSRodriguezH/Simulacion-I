```python
def experimento(n):
  #Paso 1: Identificamos las variables
  exitos = 0
  for i in range(n): #Generacion de n veces X1 y X2 (Paso 5)
    #Paso 3.-
     x1 = np.random.normal(1.5, np.sqrt(0.0016))
     x2 = np.random.normal(1.48, np.sqrt(0.0009))

     #Paso 4.-
     if x2 >= x1: #condicion de interferencia
        exitos += 1


  prob = exitos/n
   #Paso 2.- Determinamos la distribucion de probabilidad

  return prob
```

```python
import matplotlib.pyplot as plt
import numpy as np
import random as rd
```

```python
def cojinete(n):
    exitos = 0
    for i in range(n):
        x1 = np.random.normal(1.5, 0.04)
        x2 = np.random.normal(1.48, 0.03)
        if x1 > x2:
            exitos += 1
    prob = exitos / n
    return prob
```

Inciso a)

```python
p = cojinete(10000)
print(f"Probabilidad estimada de interferencia: {p:.4f}")
```

```python

#Usamos la formula del tamao de muestra de proporciones:

def tam_muestra(p, E, Z):
  n  (Z**2 * p * (1 - p)) / E**2
  return int(np.ceil(n))
```
